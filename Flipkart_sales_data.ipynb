```python
df= pd.read_excel("flipkart_sales.xlsx")
df
```

```python
import pandas as pd

df["order_date"] = pd.to_datetime(df["order_date"])
# The 'customer_rating' column likely contains numerical ratings, not dates.
# It should not be converted to datetime.
# Instead, you can ensure it's treated as a numeric type:
df["customer_rating"] = pd.to_numeric(df["customer_rating"], errors='coerce')
# errors='coerce' will replace any invalid numeric values with NaN.
df
```

```python
df = df.drop(df.columns[-1], axis=1)
df
```

#Investigating the data

```python
#plotting all variables
df.hist(figsize = (12,10))
plt.show()
```

```python
#find the total overall sales
total_revenue= df["total_sales"].sum()
print(f"Total sales:{total_revenue:,.2f}")

print("*"*30)

average_revenue= df["total_sales"].mean()
print(f"Average sales:{average_revenue:,.2f}")
```

```python
#check unique components in  categorical columns
for col in ['product_name','category','payment_method']:
  unique_values= np.unique(df[col])
  nr_values= len(unique_values)
  if len(unique_values)<= 10:
    print("the values for {} is {}.......{}".format(col, nr_values, unique_values))
  else:
    print("the number of values for feature {} is {}".format(col, nr_values))
```

```python
#find the Top 5 Best selling products
top_products= df.groupby('product_name')['quantity_sold'].sum().sort_values(ascending=False).head(5)
print("\n Top 5 best selling products")
print("*"*30)
print(top_products)

#plot a Bar Graph
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.bar(top_products.index, top_products.values, color= 'deepskyblue')
plt.xlabel("product_name")
plt.ylabel("quantity_sold")
plt.title("Top 5 Best selling products")
plt.xticks(rotation=45)
plt.show
```

```python
#Sales By Category
top_category= df.groupby('category')['total_sales'].sum().sort_values(ascending=False).head(5)
print("\n Top best category  by Revenue")
print("*"*30)
print(top_category)

#plot a Bar Graph
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.bar(top_category.index, top_category.values, color= 'deepskyblue')
plt.xlabel("category")
plt.ylabel("total_revenue")
plt.title("Category by Revenue")
plt.xticks(rotation=45)
plt.show
```

```python
# 4️ Payment Method Distribution
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="payment_method", palette="pastel")
plt.title(" Payment Method Distribution")
plt.xticks(rotation=30)
plt.show()
```

```python
#Sales Revenue By Payment Method
top_payment= df.groupby('payment_method')['total_sales'].sum().sort_values(ascending=False).head(5)
print("\n Top best revenue  by Payment Method")
print("*"*30)
print(top_payment)

#Bar Distribution
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 5))
# Changed 'palette' to 'color' and specified a single color.
# You can use a list of colors for different bars if needed.
plt.bar(top_payment.index, top_payment.values, color="deepskyblue")
plt.xlabel("payment_method")
plt.ylabel("total_revenue")
plt.title(" Payment Method Distribution")
plt.xticks(rotation=30)
plt.show()
```

```python
# Sales Trend Over Time
df.groupby(df["order_date"].dt.to_period("M"))["total_sales"].sum().plot(kind="line", marker="o", figsize=(10, 5), title="📈 Monthly Sales Trend")
plt.xlabel("month")
plt.ylabel("total_revenue")
plt.title("Sales trend  per Month")
plt.grid()
plt.show()
```
